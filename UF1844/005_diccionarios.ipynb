```python
from pprint import pprint

lorem = "Lorem ipsum dolor sit amet, consectetur adipisicing elit. Repellat quisquam odit nam et praesentium debitis non ullam placeat. "

letras = {}
for letra in lorem:
    # if letras.get(letra):
    if letra in letras:
        letras[letra] += 1
    else:
        letras[letra] = 1
        
pprint(letras)
```

```python
claves = letras.keys()

claves
```

```python
valores = letras.values()
```

```python
valores
```

```python
for c in claves:
    print(c)

```

```python
elementos = letras.items()
print(elementos)
```

```python
for c in elementos:
    print(f"Clave: {c[0]}, Valor:{c[1]}")
```

```python
otro = {
    1: 'a',
    2: 'b',
    3: 'c',
    4: 'd'
}

elem = otro.items()

salida = {}
for e in elem:
    salida[e[1]] = e[0]
    
pprint(salida)
```
