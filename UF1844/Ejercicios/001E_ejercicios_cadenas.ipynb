1. Escriba un programa que solicite al usuario que ingrese una cadena. El programa debería mostrar lo siguiente:
- (a) El número total de caracteres de la cadena
- (b) La cadena repetida 10 veces
- (c) El primer carácter de la cadena (recuerde que los índices de cadena empiezan en 0)
- (d) Los tres primeros caracteres de la cadena
- (e) Los tres últimos caracteres de la cadena
- (f) La cadena al revés
- (g) El séptimo carácter de la cadena si la cadena es lo suficientemente larga; en caso contrario, un mensaje
- (h) La cadena sin el primer y el último carácter
- (i) La cadena en mayúsculas
- (j) La cadena con cada a reemplazada por una e
- (k) La cadena con cada letra reemplazada por un espacio

```python
# Ejercicio 1

cadena = input("Introduzca una cadena:")

print(cadena)
print(f'La cadena tiene {len(cadena)} caracteres.')
print(cadena * 10)
print(cadena[0])
print(cadena[:3])
print(cadena[-3:])
print(cadena[::-1])
if len(cadena) >= 7:
    print(cadena[6])
else:
    print("Un mensaje")

print(cadena[1:-1])
print(cadena.upper())
print(cadena.replace("a", "e"))
print( " " * len(cadena))
```

2. Una forma sencilla de estimar el número de palabras en una cadena es contar el número de espacios en la cadena. <br>Escriba un programa que solicite al usuario una cadena y devuelva una estimación de cuántas palabras tiene.

```python
cadena = input("introduce una cadena:")
espacios = cadena.count(" ")
print(cadena)
print(f'Se estiman {espacios} palabras.')
print(f'Tiene {len(cadena.split())} palabras.')
```

3. A menudo se olvida cerrar los paréntesis al introducir fórmulas. <br>Escriba un programa que solicite al usuario que introduzca una fórmula e imprima si esta tiene el mismo número de paréntesis de apertura y cierre.

```python
formula = input("Introduce una fórmula con paréntesis: ")
abiertos = formula.count("(")
cerrados = formula.count(")")
if abiertos != cerrados:
    print(f"La fórmula tiene {abiertos} paréntesis abiertos y {cerrados} peréntesis cerrados. Te faltan {abiertos-cerrados}")
else:
    print(f"La formula tiene la misma cantidad de paréntesis abiertos y cerrados.")
```

4. Escriba un programa que solicite al usuario que introduzca una palabra e imprima si contiene vocales.

```python
palabra = input("Introduce una palabra:")
if any(letra in "aeiouAEIOU" for letra in palabra):
    print(f"La palabra {palabra} contiene vocales.")
else:
    print(f"La palabra {palabra} no contiene vocales.")
```

7. Escriba un programa que solicite al usuario que ingrese una palabra y determine si es un palíndromo.<br>
Un palíndromo es una palabra que se lee igual de derecha a izquierda.

```python
palindromo = input("Introduce una cadena: ")
palindromo = palindromo.lower()
if palindromo == palindromo[::-1]:
    print(f"La palabra {palindromo} es un palíndromo.")
else:
    print(f"La palabra {palindromo} no es un palíndromo.")
```

9. Solicite al usuario un número e imprima lo siguiente, donde el patrón termina en el número ingresado por el usuario. 

1<br>
&nbsp;2<br>
&nbsp;&nbsp;3<br>
&nbsp;&nbsp;&nbsp;4

```python
numero = input("Introduce un número:")
for n in range(1, int(numero) + 1):
    print( f'{" " * (n - 1)}{n}')
```
