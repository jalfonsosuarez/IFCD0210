4. Escriba una función que devuelva la raíz digital de un entero n.<br>
La raíz digital de un número n se obtiene de la siguiente manera: <br>
Sume los dígitos de n para obtener un nuevo número. <br>
Sume los dígitos de ese para obtener otro número nuevo. <br>
Continúe haciendo esto hasta obtener un número que solo tenga un dígito. <br>
Ese número es la raíz digital. <br><br>

Por ejemplo, si n = 45893, sumamos los dígitos para obtener 4 + 5 + 8 + 9 + 3 = 29. <br>
Luego sumamos los dígitos de 29 para obtener 2 + 9 = 11. <br>
Luego sumamos los dígitos de 11 para obtener 1 + 1 = 2. <br>
Como 2 tiene solo un dígito, 2 es nuestra raíz digital.
