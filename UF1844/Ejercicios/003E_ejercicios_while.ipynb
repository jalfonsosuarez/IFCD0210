1. El código a continuación imprime los números del 1 al 50. <br>
Reescribe el código usando un bucle while para lograr lo mismo.
```python
for i in range(1,51):
    print(i)
```


````python
contador = 1
while contador <= 51:
    print(contador)
    contador += 1
````

3. Un buen programa se asegurará de que los datos que ingresan los usuarios sean válidos. <br>
Escribe un programa que solicite al usuario un peso y lo convierta de kilogramos a libras. <br>
Siempre que el usuario ingrese un peso menor a 0, el programa le indicará que su entrada no es válida y le solicitará nuevamente que ingrese un peso. <br>


4. Escribe un programa que solicite al usuario una contraseña. <br>
Si el usuario ingresa la contraseña correcta, el programa le indicará que ha iniciado sesión en el sistema. <br>
De lo contrario, le pedirá que vuelva a ingresar la contraseña. <br>
El usuario solo tendrá cinco intentos para ingresar la contraseña, tras lo cual el programa le indicará que ha sido expulsado del sistema.


````python
MAX_INTENTOS = 3
pwd = ""
i = 0
while i < MAX_INTENTOS:
    pwd = input("Introduce la contraseña:")
    if pwd == "contraseña":
        print("Contraseña correcta")
        break;
    else:
        i += 1
        print("Contraseña incorrecta")

if i == MAX_INTENTOS:
    print("Su cuenta ha sido borrada. jajajajajajaja")
````

8. El MCD (máximo común divisor) de dos números es el número más grande entre el cual ambos son divisibles. <br>
Por ejemplo, mcd(18, 42) es 6 porque el número más grande entre el cual 18 y 42 son divisibles es 6. <br>
Escriba un programa que solicite al usuario dos números y calcule su mcd.<br>
A continuación, se muestra una forma de calcular el MCD, llamada Algoritmo de Euclides.
- Primero, calcule el resto de dividir el número mayor entre el menor.
- Luego, reemplace el número mayor con el menor y este con el resto.
- Repita este proceso hasta que el número menor sea 0. <br>
El MCD es el último valor del número mayor.
