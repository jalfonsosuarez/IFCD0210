## Bucles en python

```python
for i in range(10):
    print(i)
```

```python
for i in range(1, 25, 3):
    print(i)
```

```python
for i in range(25, 1, -3):
    print(i)
```

```python
lista = [ 1,2,3,4,5,6,7,8,9,0]

for num in lista:
    print(num)
```

```python
cadena = "1234567890"

for item in cadena:
    print(item)
```

```python
lista1 = [1,2,3]
lista2 = [4,5,6]
lista_total = lista1 + lista2
lista_total
```

```python
for n in lista_total:
    if n % 2 == 0:
        print(f"{n} es par")
```
