```python
def varios():
    return 1, 'a', 2, 'b'

x = varios()
print(x)

x1, x2, x3, x4 = varios()
print(x1, x2, x3, x4)
```

```python
def saludar(nombre: str) -> str:
    """
    saludar: emite un saludo
    
    Parámetros: nombre de tipo cadena
    
    Devuelve: Cadena con el saludo
    """
    return f'Hola {nombre}'

saludo = saludar("Jose")
print(saludo)
```

```python
def mostrar_config(**kwargs):
    for k, v in kwargs.items():
        print(k, v)

mostrar_config(user="Jose", IP="192.168.1.33", port=9080, height=1.75)
```

```python
d = {
    'usr': 'Jose',
    'ip': '192.168.1.40',
    'puerto': '8080'
}

mostrar_config(**d)
```

```python
def log(mensaje, *args, **kwargs):
    nivel = kwargs.get('nivel', 'DEBUG')
    prefijo = kwargs.get('prefijo', '')
    msj_formato = f'[{nivel}]: {prefijo} - {mensaje}'
    return msj_formato

log('Kernel panic', nivel='ERROR', prefijo = 'SYSTEM')
```

```python
def http_get(url, **kwargs):
    timeout = kwargs.get('timeout', 5)
    headers = kwargs.get('headers', {})
    params = kwargs.get('params', {})
    return {
        'url': url,
        'timeout': timeout,
        'headers': headers,
        'params': params
    }
    
http_get('http://www.data.com', timeout=30, params={'page':3})
```
