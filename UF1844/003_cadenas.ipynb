## Entrada de datos por teclado

```python
dato = input("Dame un múmero: ")
print(int(dato) + 10)
```

```python
num = input("Dame un numero, pls: ")
if num.isdigit():
    n = int(num)
    if n > 0:
        print( f"{n} es mayor que 0")
    elif n == 0:
        print(f"{n} es 0")
    else:
        print(f'{n} es menor que 0')
else:
    print("¡¡¡Te dije que me dieras un número, merluzo!!!")
```

```python
cadena = "texto de prueba"

letra = input("Dame una letra:")

if letra in cadena:
    print(f"La letra {letra} está en la cadena '{cadena}'")
else:
    print(f"La letra {letra} no está en la cadena '{cadena}'")
    
```
