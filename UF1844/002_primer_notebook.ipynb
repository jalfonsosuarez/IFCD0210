Calculo del índice de masa corporal.

```python
peso = 100
altura = 1.74
imc = peso / altura ** 2
mensaje = "El IMC para un peso de " + str(peso) + "Kg y una altura de " + str(altura) + " m es de " + str(imc)
print(mensaje)
if imc < 18:
    print("El IMC es bajo")
elif imc < 25:
    print("El IMC es normal")
else:
    print("El IMC es alto: sobrepeso")
```
